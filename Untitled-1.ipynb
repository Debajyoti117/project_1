```python
import pygame
import random

# Initialize pygame
pygame.init()

# Screen settings
width = 600
height = 500
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Car Racing Game")

# Colors
white = (255, 255, 255)
black = (0, 0, 0)
red = (200, 0, 0)

# Car settings
car_width = 80
car_height = 180
car_x = width // 2
car_y = height - 120
car_speed = 10

# Enemy car
enemy_width = 50
enemy_height = 100
enemy_x = random.randint(0, width - enemy_width)
enemy_y = -100
enemy_speed = 10

clock = pygame.time.Clock()
score = 0

running = True
while running:

    screen.fill(white)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    keys = pygame.key.get_pressed()

    if keys[pygame.K_LEFT]:
        car_x -= car_speed
    if keys[pygame.K_RIGHT]:
        car_x += car_speed

    # Enemy movement
    enemy_y += enemy_speed

    if enemy_y > height:
        enemy_y = -100
        enemy_x = random.randint(0, width - enemy_width)
        score += 1

    # Collision detection
    if (car_y < enemy_y + enemy_height and
        car_y + car_height > enemy_y and
        car_x < enemy_x + enemy_width and
        car_x + car_width > enemy_x):
        print("Game Over! Score:", score)
        running = False

    # Draw car
    pygame.draw.rect(screen, black, (car_x, car_y, car_width, car_height))

    # Draw enemy
    pygame.draw.rect(screen, red, (enemy_x, enemy_y, enemy_width, enemy_height))

    pygame.display.update()
    clock.tick(100)

pygame.quit()
```
